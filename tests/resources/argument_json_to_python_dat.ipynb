The argument cases that are used to initialize the agent's argumentation CBRs are stored in JSON files. In order to ease
its us the agents read directly from .dat files that contain the cases directly as Python objects. This notebook allows
the user to get an easy translation from the JSON files to the desired .dat files

The required imports

```python
from os.path import isfile, abspath, join as join_path, splitext
from os import listdir

from loguru import logger

from tests.resources.json_to_python_utils import translate_argument_cases as translate
```

Stabilising the folders

```python
json_cases_dir = abspath("json/argument_cases/partArgInc")
logger.info("JSON sources directory set to: {}".format(json_cases_dir))
python_cases_dir = abspath("data/argument_cases/partArgInc")
logger.info("Data destination directory set to: {}".format(python_cases_dir))
```

Every JSON file in the source folder is translated and stored as Python objects in the
destination directory

```python
logger.info("Starting the directory walk in: {}".format(python_cases_dir))

json_files = [f for f in listdir(json_cases_dir) if isfile(join_path(json_cases_dir, f)) and
              f.endswith('.json')]
for file in json_files:
    fullname = join_path(json_cases_dir, file)
    dest_name = splitext(join_path(python_cases_dir, file))[0]
    dest_name += ".dat"
    translate(fullname, dest_name, verbose_test=True)
logger.info("Directory translation finished")
```
